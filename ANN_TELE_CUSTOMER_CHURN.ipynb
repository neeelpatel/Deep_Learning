```python
#convert categorical value into numeric
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
data[['gender','Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling']] = data[['gender','Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling']].apply(labelencoder.fit_transform)

```

```python
#scalling
from sklearn.preprocessing import MinMaxScaler
model = MinMaxScaler()
data[['tenure','MonthlyCharges','TotalCharges']]=model.fit_transform(data[['tenure','MonthlyCharges','TotalCharges']])
```

```python
data
```

```python
#dependent and independent data
x=data.drop(['Churn'],axis='columns')
y=data['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=5)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
#create ANN ,initializing model
classifier = keras.Sequential()
#adding inputlayer,here we have 24 columns as a feature
classifier.add(Dense(units=20,activation="relu",input_shape=(23,)))
#add secound layer
classifier.add(Dense(units=15,activation='relu'))
#add another hidden layer
classifier.add(Dense(units=10,activation="relu"))
classifier.add(Dense(units=1,activation="sigmoid"))





```

```python
#train model
classifier.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
```

```python
model = classifier.fit(x_train,y_train,validation_split=0.33,epochs=100)
```

```python
#evalute model 
classifier.evaluate(x_test,y_test)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

here, problem is we get two dimension array so we need to convert it into one dimenstion array.

and in y_test we have o and 1 only so we need to convert y_pred into one dimention

if y_pred > 0.5 then convert it into 1 other wise convert it into 0


```python
yp= []
for i in y_pred:
    if i > 0.5:
        yp.append(1)
    else:
        yp.append(0)
```
